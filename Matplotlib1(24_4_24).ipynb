```python
#Font properties adjustments
# first create a dictonary and assign them to fontdict
import matplotlib.pyplot as plt
import numpy as np
ft = {"family" : "arial" , "color" : "green" , 'size' : 15}
f1 = {"family" : "cambria", "color" : "blue" , 'size' : 15}
f2 = {"family" : "cambria", "color" : "blue" , 'size' : 15}
x = np.arange(0,12,2)
y = x ** 10
plt.title("Multiples Of 10", fontdict = ft, horizontalalignment= "right")
plt.xlabel("X-axis",fontdict = f1,loc = "right")
plt.ylabel("Y-axis",fontdict = f2,loc = "top")
plt.plot(x,y,marker = "o",ms = 10, mfc = "green", mec = "green")
plt.plot(y, ls= "dotted",c = "r",lw =3)
plt.show()
```

```python
#scatter plot with different colors and sizes for different elements
import matplotlib.pyplot as plt
import random
import numpy as np
x = np.arange(1,5,1)
y = np.arange(21,25,1)
print(x,y)
colors = [20,25,50,34]
sizes = [50,15,90,18]

plt.scatter(x,y,c =colors ,cmap = 'rainbow',s = sizes,alpha = 0.9)
plt.colorbar()
plt.show()
```

```python
# bar and barh graphs
import matplotlib.pyplot as plt
x = ['2018','2019','2020','2021','2022',"2023","2024"]
y = [300,350,320,450,250,260,150]
c = ["red","green","orange","yellow","blue"]
plt.bar(x,y,color = c,width = 0.6)
plt.title("MGIT yearly placements")
plt.xlabel("year")
plt.ylabel("no.of.placements")
plt.show()
plt.barh(x,y,color = c,height = 0.6)
plt.title("MGIT yearly placements",loc = 'left')
plt.xlabel("no.of.placements")
plt.ylabel("year")
plt.show()
```

```python
#using stacked br chart
import matplotlib.pyplot as plt
years = ['2018','2019','2020','2021','2022']

civils = [55,68,75,54,45]
mtech = [45,51,24,64,61]
ms = [34,45,15,34,51]
#mba = [64,34,75,50,39]
plt.title("different carriers using stacked bar charts")
plt.bar(years,civils,width = .6,color ="black",label = "civils")
plt.bar(years,mtech,width = .6,color ="red",label = "mtech")
plt.bar(years,ms,width = .6,color ="green",label  = "ms")
plt.ylim(0,100)
plt.legend()
plt.show()

```
