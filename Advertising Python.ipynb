## now the data is clean and duplicate free ,so we can now plot graphs to check the replationships

```python
# Distribution Plots
for i, col in enumerate(['TV', 'Radio', 'Newspaper', 'Sales']):
    plt.figure(figsize=(10, 6)) # ensures each graph is plotted one by one.
    plt.subplot(2, 2, i+1)
    sns.histplot(a_df[col], kde=True) #kernal density estimante -> curve
    plt.title(f'Distribution of {col}')
plt.tight_layout()
plt.show()
```
