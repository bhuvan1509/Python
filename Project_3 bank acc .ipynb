```python
#Bank account creation
# using dictonaries
details = {1 : {'name' : "sdfs",'age' : 5,'aadhar' : 5,'pan' : 'df','phno' : "asdf",'village' : "asdf"} , 
          2 : {'name' : "sdfs",'age' : 5,'aadhar' : 5,'pan' : 'df','phno' : "asdf",'village' : "asdf"} }
t = 0
ano = 2
ifsc = "SBIN545645"
def take_details(ano):
    details[ano]['name']= input("enter your name:")
    details[ano]['age'] = input("enter your age:")
    details[ano]['aadhar'] = input("enter your aadhar number:")
    details[ano]['pan'] = input("enter your PAN:")
    details[ano]['phno'] = input("enter your phone number:")
    details[ano]['village'] = input("enter your village name:")
    
    
def acc_open():
    global ano
    print("which account do you want : ZERO or Salaried ")
    type = input()
    
    if type == "zero" :
        take_details(ano)
        print("since zero , no deposit is required")
        ano += 5  # random acc no
        print(f"yout account number is {ano}")
        print(f"your ifsc is {ifsc}")
        t_options()
        
        
    elif type == "salaried" :
        take_details(ano)
        print("how much amount are you depositing(min = 1000) ")
        deposit = int(input("deposit:"))
        ano += 5  # random acc no
        print(f"yout account number is {ano}")
        print(f"your ifsc is {ifsc}")
        t_options()
              
    else :
        print("please ,enter a valid option!!!!")
t = 0        
def t_options():
    global t
    global ano
    print("""
    0 : users list
    1 : add money / Deposit
    2 : Balance
    3 : withdraw    
    4 or more to exit
    """)
    opt = int(input("enter transaction: "))
    if opt == 1 :
        print("amount")
        amount = int(input("enter amount to deposit :"))
        t = t + amount
        print(f"{amount} is successfully deposited to your account {ano}")
        t_options()
        
    elif opt == 2:
        print(f"your total balance is :{t}")  
        t_options()
        
    elif opt == 3:
        a = int(input("enter amount to withdraw:"))
        t -= a
        print(f"RS {a} withdraw is successfull ")
        t_options()  
    elif opt == 0 :
        # table header
        print("| {:^5} | {:^10} | {:^5} | {:^10} | {:^10} | {:^15} |".format("ID", "Name", "Age", "Aadhar", "PAN", "Phone Number", "Village"))

        # separator
        print("|{:-<7}|{:-<13}|{:-<8}|{:-<12}|{:-<11}|{:-<16}|".format("", "", "", "", "", ""))
        
        # details
        for key, value in details.items():
            print("| {:^5} | {:<10} | {:^5} | {:^10} | {:<10} | {:<15} |".format(key, value['name'], value['age'], value['aadhar'], value['pan'], value['phno'], value['village']))

        
    
    
    
acc_open()
```
